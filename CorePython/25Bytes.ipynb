**Creating byte variable and store byte data**

```python
b1=bytes([97,98,99,100,101])
print(b1)
print(type(b1))
```

```python
b2=bytes('Hello', encoding='utf-8')
print(b2)
print(type(b2))
```

```python
b3=b'Python'
print(b3)
print(type(b3))
```

```python
b4=bytes(5)
print(b4)
```

```python
b5=bytes.fromhex('536F68616D')
print(b5)
```

**bytearray**

```python
ba1=bytearray([65,66,67,68,69])
print(ba1)
```

```python
ba2=bytearray(b'Praffull')
print(ba2)
ba2[0]=83
print(ba2)
```

```python
ba2.insert(0,35)
print(ba2)
```
