```python
#base class
class Praffull:
  def calcsquare(self,n):
    print(f"square is {n*n}")
```

```python
#derived class
class Sharayu(Praffull):
  def calcsum(self,x,y):
    print(f"sum is {x+y}")
```

```python
obj=Sharayu()
```

```python
obj.calcsum(12,34)
```

```python
obj.calcsquare(13)
```
