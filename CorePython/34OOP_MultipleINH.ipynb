```python
class One:
  def calcsquare(self,n):
    print(f"square is {n**2}")
```

```python
class Two:
  def calcsum(self,x,y):
    print(f"sum is {x+y}")
```

```python
class Three(One,Two):
  def calcdiscount(self,amount):
    print(f"discount will be {amount*10/100}")
```

```python
obj=Three()
```

```python
obj.calcdiscount(5400)
```

```python
obj.calcsum(45,67)
```

```python
obj.calcsquare(15)
```
