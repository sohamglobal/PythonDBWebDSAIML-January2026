```python
from abc import ABC, abstractmethod

class Payment(ABC):
  @abstractmethod     # decorator
  def pay(self):
    pass
```

```python
class UPIPayment(Payment):
  def pay(self):   # taken from the base
    print("bill paid using PhonePe")
  def summary(self):  # own function
    print('summary of transactions')
```

```python
obj=UPIPayment()
```

```python
obj.pay()
```

```python
obj.summary()
```
