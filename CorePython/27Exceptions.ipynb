Exception Handling

```python
try:
  n=int(input('Enter a number : '))
except:
  print("Invalid input...please enter a number")
  n=0

sq=n**2
print(f"square of {n} is {sq}")
```

```python
custnm=input('Enter customer name : ')

try:
  purchamt=float(input('Enter purchase amount : '))
except ValueError:
  print('invalid purchase amount')
  purchamt=0

discount=purchamt*10/100
bill=purchamt-discount

print(f"{custnm}, your net bill is {bill}")
```

```python
try:
  x=int(input('Enter first number '))
  y=int(input('Enter second number '))
  res=x/y
except ValueError:
  res=0
  print('Invalid input')
except ZeroDivisionError:
  res=x
  print('cant divide by zero')

print(f'result of division is {res}')
```

```python
try:
  x=int(input('Enter first number '))
  y=int(input('Enter second number '))
  res=x/y
except:
  res=0
  print('Invalid input or divided by zero')

print(f'result of division is {res}')
```
