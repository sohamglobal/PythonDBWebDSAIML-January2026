variable positional arguments

```python
def congrats(*names):
  print(names)
  for nm in names:
    print(f"congrats {nm}")
```

```python
congrats('jos buttler')
```

```python
congrats('joe root','steve smith')
```

```python
congrats('steffi graff','emily blunt','dominik szoboszlai','florian wirtz')
```

```python
def total(*args):
  print(f"sum of values is {sum(args)}")
  print(f"largest value is {max(args)}")
```

```python
total(2,5)
```

```python
total(23,12,76,49,16)
```

variable keyword arguments

```python
def display(**kwargs):
  for key,value in kwargs.items():
    print(f"Key : {key}, Value : {value}")
```

```python
display(name="praffull",city="berlin",club="chelsea")
```

```python
display(player="virgil van dijk",country="netherlands",position="defender",club="liverpool")
```
