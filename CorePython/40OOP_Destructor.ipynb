```python
class Employee:
  def __init__(self):
    print("welcome to SohamGlobal")
  def calcsalary(self,basic):
    allowances=int(basic*45/100)
    tax=int(basic*5/100)
    netsalary=basic+allowances-tax
    print(f"Net Salary : {netsalary}")
  def __del__(self):
    print('Object destroyed from the memory')
```

```python
e=Employee()
```

```python
e.calcsalary(41500)
```
