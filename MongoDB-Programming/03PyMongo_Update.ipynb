```python
db1=client["spiderdb"]
```

```python
coll_workers=db1["workers"]
```

```python
nm=input('Enter worker name : ')
condition={}
condition['name']=nm
print(condition)
```

```python
loc=input('Enter new location : ')
change={}
change['location']=loc
print(change)

upd={}
upd['$set']=change
print(upd)
```

```python
cnt=coll_workers.update_one(condition,upd)
print(cnt.modified_count)
if cnt.modified_count>0:
  print('worker location changed successfully')
else:
  print('worker not found')
```

```python
db2=client["sohamdb"]
```

```python
coll_users=db2["users"]
```

```python
uid=input('Enter userID : ')
condition={}
condition['userid']=uid
print(condition)
```

```python
nps=input('Enter new password : ')
change={}
change['password']=nps
print(change)

upd={}
upd['$set']=change
print(upd)
```
