```python
class Product:
  quantity=0
  purchcost=0
  def __init__(self,quantity,purchcost):
    self.quantity=quantity
    self.purchcost=purchcost
  def showstock(self):
    print(f"Quantity : {self.quantity}")
    print(f"Purchase Cost : {self.purchcost}")
  def __add__(self,obj):
    q=self.quantity+obj.quantity
    c=self.purchcost+obj.purchcost
    return Product(q,c)
```

```python
dell=Product(9,641500)
dell.showstock()
```

```python
hp=Product(14,955300)
hp.showstock()
```

__add__,__sub__,__mul__,__truediv__,__mod__ are the functions we can add to the class to overload operators +,-,*,/,% resp

```python
total=dell+hp
total.showstock()
```
