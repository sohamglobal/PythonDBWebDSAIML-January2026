Slicing without copying

```python
slogan=bytearray(b'Technology is power')
print(slogan)
```

```python
slice1=slogan[:7]
print(f"regular slicing : {slice1}")
```

```python
view=memoryview(slogan)
vslice=view[2:8]
print(f"Memoryview slice : {bytes(vslice)}")
```
